我们可以用2*1 的小矩形横着或者竖着去覆盖更大的矩形。请问用 n个2*1的小矩形无重叠地覆盖一个2*n的大矩形，总共有多少种方法？

比如n=3时，2*3的矩形块有3种覆盖方法：


# 分析
2*1 覆盖 2*n 
可以分为横覆盖与竖覆盖
n=n时 若最后一个横覆盖，最后第二个必定也为横覆盖，因此只需考虑n-2次；最后一个竖覆盖，只需考虑n-1次

$f(n)=f(n-2)+f(n-1)$
因此和斐波那契数列相似

```python
# -*- coding:utf-8 -*-
class Solution:
    def rectCover(self, number):
        # write code here
        a, b = 1, 2
        if number <= 1:
            return number
        while number > 1:
            c = a + b
            a = b
            b = c
            number -= 1
        return a

res = Solution()
res.rectCover(1)
```
